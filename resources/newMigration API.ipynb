```python
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
url = "https://open.alberta.ca/dataset/eece1571-4b82-40be-8a11-4627db1a9561/resource/2acfe7ec-6c54-48ea-bcb8-6aa72fd64a7a/download/rows"
response = requests.get(url).json()
```

```python
#Parse API
net_migration =[]
time_period = []
for index in range(0,831,3):
    net_migration.append(response['data'][index][12])

for index in range(0,831,3):
    time_period.append((response['data'][index][8].split("-",1)[0]))

```

```python
#Create Dataframe
migration_df = pd.DataFrame({"Date": time_period,"National Net Migration": net_migration})

migration_df["National Net Migration"] = pd.to_numeric(migration_df["National Net Migration"])

```

```python
#Club quarterly data to get annual data
grouped_migration = migration_df.groupby(['Date'])['National Net Migration'].sum().reset_index()
grouped_migration
```

```python
# Push the remade DataFrame to a new CSV file
grouped_migration.to_csv('newnationalmigration_clean.csv',
                  index=False, header=True)
```
