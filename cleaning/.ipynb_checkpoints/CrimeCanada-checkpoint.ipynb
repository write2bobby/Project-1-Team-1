```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '../resources/35100131.csv'
data = pd.read_csv(path)
data
```

```python
all_offences_canada = data[data['Type of offence'] == 'All offences, total']
all_offences_canada = all_offences_canada[all_offences_canada['GEO'] == 'Canada'][['REF_DATE', 'VALUE']]
all_offences_canada
```

```python
fig, axes = plt.subplots()
axes.plot(all_offences_canada['REF_DATE'], all_offences_canada['VALUE'])
axes.set(title = 'Total Crime in Canada (1965 - 2000)', xlabel='Year', ylabel='Incidences')
```

```python
all_offences_canada.columns = ['Date', 'National Crime Incidences']
all_offences_canada.to_csv('../Resources/TotalCrimeCanada.csv', index_label = False)
```
